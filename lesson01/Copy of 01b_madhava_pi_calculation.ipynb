This notebook is designed to be a live demonstration of Python fundamentals applied to calculating Ï€ using the [Madhava-Leibniz series](https://en.wikipedia.org/wiki/Approximations_of_%CF%80#Middle_Ages)

The approximation is simple:

$$ \pi \approx 4 \cdot \left(1 - \frac{1}{3} + \frac{1}{5} - \frac{1}{7} + \frac{1}{9} \ldots -(-1)^n \frac{1}{2n - 1}  \right) $$

```python
# pi starts out at 0. initialize this
pi = 0

#define n
n=1

#put n in approximation formula
x=((-1)**n)/(2*n-1)

#define pi
pi=4*(x)

#print pi
print(pi)
```

```python
# Travis solution
# pi starts out at 0. initialize this
pi = 0

# set number of iterations
N = 10000

# initialize empty list of pi vals
pi_values = []

# loop from 0 up to number of iterations
for n in range(1,N):
  # calculate the current term in the pi approximation
  current_term = 4*( -(-1)**n * 1/(2*n - 1))

  #add to current value of pi
  pi = pi + current_term

  #append pi (what is this?)
  pi_values.append(pi)

print(pi)
```

```python
"""Plot terms of pi as function of iteration"""
# import plotting library (you will need to do this often)
import matplotlib.pyplot as plt

plt.plot(pi_values)
plt.xscale('log')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Madhava")
```
