In this group exercise, you will mimic what we did with Ï€: you will write code to calculate $e$ using the following series approximation (from [wikipedia](https://en.wikipedia.org/wiki/List_of_representations_of_e#As_an_infinite_series)):


$$ e = \sum\limits_{k=0}^{\infty} \frac{1}{k!} $$

where $k!$ represents 'k factorial': $k! = 1 \cdot 2 \cdot 3 \cdot \ldots \cdot k$


```python
"""trying on my own"""
#initialize e (or k?)
e=0

# set number of iterations
K = 10000

# initialize empty list of k vals
k_values = []

# loop from 0 up to number of iterations
for k in range(0,K):
  # calculate the current term in the k approximation
  current_term = k

  #add to current value of k
  k = k * current_term +1

  #append k (what is this?)
  k_values.append(k)

#
print(k)
```

```python
"""not using math factorial function"""
#initialize e (or k?)
e=0

# set number of iterations
K = 100

# initialize empty list of e vals
e_values = []

# loop from 0 up to number of iterations
for k in range(0,K):

#compute factorial. first define
  fact = 1

  for i in range(1,k+1):
    fact = i*fact


  # calculate the current term in the factorial approximation
  current_term = 1/fact

  #start adding the "e"s
  e = e + current_term

  #append e
  e_values.append(e)

#
print(e)
```
