There is a useful concept that we haven't touched on yet: complex numbers in Python.  You could, for example, define the imaginary number $i = \sqrt{-1}$ in Python as `i = 1.0j`, where the `j` after the number indicates that the number is the imaginary part of a complex number.

```python
"""demonstration of a function"""

def double(number):
  """doubles the number"""
  # multiply it by 2
  doubled_num = 2*number

  return doubled_num

#test. double 3.21 is 6.42
double(3.21)
```

```python
"""creating cos(z)"""
#using formula cos(z)=1/2(e**(i*z)+(e**(-i*z)))
#define e from exercise 1c
e = 2.7182818284590455

#define pi from exercise 1b

pi = 3.1416926635905345

#define i
i = 1.0j

#define cos as a function with angle z in radians
def cos(z):
  cos_z = 1/2*((e**(i*z))+(e**(-i*z)))
  return cos_z

#convert degrees into radians
def deg2rad(deg):
  rad = deg*pi/180
  return rad

#test. at 180 degrees, cos is -1
cos(deg2rad(180))
```

```python
"""creating exp(w)"""
#using formula exp(w) = e**(w*i)

# e, i, and pi are defined in the previous cell

#define exp as a function with w
def exp(w):
  exp_w = e**(w)
  return exp_w

#test. e^w should return e*e if w=2, which is about 7.389
exp(2)
```

```python
"""creating sin(y)"""
#using formula sin(y) = 1/2(e**(i*z)+(e**(-i*z)))

# e, i, and pi are defined in the first cell, as well as degrees to radians

#define sin as a function with y
def sin(y):
  sin_y = 1/2*(e**(i*y)-(e**(-i*y)))
  return sin_y

#test. at 90 degrees, sin is 1
sin(deg2rad(90))
```

```python
"""creating tan(x)"""
#using formula tan(x) = sin(x)/cos(x)

# functions sin and cos are defined in previous cells, as well as degrees to radians

#define tan as a function with x
def tan(x):
  tan_x = (sin(x))/(cos(x))
  return tan_x

#test. at 45 degrees, tan is 1
tan(deg2rad(45))
```
