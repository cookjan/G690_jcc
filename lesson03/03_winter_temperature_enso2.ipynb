This notebook examines the relationship between winter minimum temperatures in southern central IN and El Nino.

**Goal:** examine the relationship between El Nino and minimum winter (Dec-Jan-Feb) temperatures in Bloomington, IN

**Method:**

 * obtain minimum temperature data from [NOAA NClimDiv](https://www.ncei.noaa.gov/access/metadata/landing-page/bin/iso?id=gov.noaa.ncdc:C00005), using the *NCEI Direct Download* method
 * read the dataset using `pandas`, filter out data for southern central Indiana
 * convert the data to timeseries format
 * obtain the [ENSO longitude index](https://cascade.lbl.gov/enso-longitude-index-eli/) Excel file and read with `pandas`; convert to timeseries format
 * make the ENSO and NClimDiv timeseries align (same start and end dates, same period - DJF, etc)
 * plot both timeseries
 * plot a scatter plot
 * do linear regression and correlation analyses

For this solution notebook, I will be using data directly from NOAA instead of downloading it, since `pandas` has the ability to open data from a URL.

# Part 1: Getting the NClimDiv data

```python
"""import libraries"""
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

```python
"""read the tmin data file"""
#filename = r"C:\Users\janel\OneDrive\Documents\IU grad\Classes F23\EAS 690 Advanced Data Science\G690_Test\lesson03\climdiv-tmindv-v1.0.0-20230804.txt"
#with panda, you can copy the url
file_url = "https://www.ncei.noaa.gov/pub/data/cirs/climdiv/climdiv-tmindv-v1.0.0-20230907"
# updated URL from https://www.ncei.noaa.gov/pub/data/cirs/climdiv/

#file_path = r'C:\Users\janel\OneDrive\Documents\IU grad\Classes F23\EAS 690 Advanced Data Science\G690_Test\lesson03\climdiv-tmindv-v1.0.0-20230804.txt'

# set the header manually
header = ["code"]
header = header + list(range(61,73))
#header = None

#read in data file
climdiv_df = pd.read_csv(file_url, header = None, names = header, delimiter = "\s+", na_values = -99.9)
climdiv_df
```

```python
"""filter out state code"""
#sets state and division code to south centeral indiana 
statecode = "12"
divcode = "08"
state_div = statecode + divcode

#get the rows that are in the requested climate division
# code is "astype" string because "startswith" funciton only works on strings
in_div = climdiv_df['code'].astype(str).str.startswith(state_div)

#select only those rows
scin_df = climdiv_df[in_div]

scin_df
```

```python
climdiv_df.min()

climdiv_df['code']
```

```python
"""convert data to timeseries"""
#convert to numpy array
scin_as_numpy = scin_df.values #.values will give you an array
#pull out only the temp data
scin_temp_np = scin_as_numpy[:,1:13]

#scin_temp_np.shape #this gives us dimensions of the array

#convert matric into a vector
scin_temps = scin_temp_np.ravel()

#plt.plot(scin_temps)

#make dates for each index
scin_dates = pd.date_range(start = '1895-01-1', end = '2024-01-01', freq = "1M")

#create a new dataframe for the timeseries
dataframe_dict = {'date' : scin_dates, 'temperature' : scin_temps}

scin_ts = pd.DataFrame(dataframe_dict)
scin_ts.plot('date', 'temperature')
#scin_ts = pd.DataFrame()

```

```python
#test_dict = {'arbitrary string': 1, 'mary' : 3.14159265}

#test_dict['arbitrary string']
```

```python
"""select average over winger times (DJF)"""

#select only winter months
date_month_values = scin_ts['date'].dt.month #get the month values for each date
in_winter = date_month_values.isin([1,2,3])
#use boolean indexing to select only winter months
scin_ts_winter = scin_ts[in_winter]

#group by year and average over the winter months
year = scin_ts_winter['date'].dt.year #pull out year for each date
scin_ts_winter_avg = scin_ts_winter.groupby(year).mean()

#plot the time series
fig, ax = plt.subplots()
scin_ts_winter_avg.plot(ax = ax)
ax.set_ylabel("Temperature (F)")
plt.show()
print(scin_ts_winter_avg)
```

# part 2: reading and opening el nino data

```python
# set URL to ELI file
eli_url = "https://portal.nersc.gov/archive/home/projects/cascade/www/ELI/ELI_ERSSTv5_1854.01-2020.05.xlsx"

# read in the data
eli_df = pd.read_excel(eli_url, sheet_name = "ELI_ERSSTv5_1854.01-2020.02", header = 0, skiprows = 0)
eli_df #should return file once you have URL and SHEET NAME!
```

hint: use merge() method in pandas to get ELI and NOAA data to align

```python
#convert to timeseries format
#from before: 
"""convert data to timeseries"""
#convert to numpy array
eli_as_numpy = eli_df.values #.values will give you an array


#pull out only the ELI (cut out month)
eli_extract_np = eli_as_numpy[:,1:-1]

#eli_extract_np.shape #this gives us dimensions of the array

#convert matrix into a vector. Order F (column major)
eli_extract = eli_extract_np.ravel("F")

#plt.plot(eli_extract) #see plot so far

#make dates for each index. Cut off unfinished year of 2020
eli_dates = pd.date_range(start = '1854-01-1', end = '2020-01-01', freq = "1M")

#create a new dataframe for the timeseries
dataframe_eli = pd.DataFrame({'date' : eli_dates, 'ELI' : eli_extract})
dataframe_eli
```

```python
""" Select and average over winter times (DJF). """

# select only the winter months
date_month_eli = dataframe_eli['date'].dt.month # get the month values for each date
in_winter_eli = date_month_eli.isin([12,1,2])
# use boolean indexing to select only the winter months
eli_winter = dataframe_eli[in_winter_eli]

# group by year and average over the winter months
Eyear = eli_winter['date'].dt.year # pull out the year for each date
ELI_winter_avg = eli_winter["ELI"].groupby(Eyear).mean()

#remake the dataframe
df_eli_wAvg = ELI_winter_avg.to_frame()
```

```python
#align both timeseries using merge and dates

eliTemp = pd.merge(df_eli_wAvg, scin_ts_winter_avg, on = 'date')
eliTemp
```

```python
#plot both time series
fig, ax = plt.subplots()
eliTemp.plot(ax = ax)
ax.set_ylabel("Value")
plt.show()

```

```python
#make scatterplot of the variables
plt.scatter(eliTemp["ELI"], eliTemp["temperature"])
plt.ylabel("Temperature (F)")
plt.xlabel("ELI")
plt.show
```

```python
"""Perform linear regression and correlation analyses"""

#calculate regession line
m, b = np.polyfit(eliTemp["ELI"], eliTemp["temperature"], 1)
print("Temperature (F) in South Central Indiana = ", m, "ELI +", b)

#calculate correlation coefficient
ET_corr = eliTemp.corr()
print("Correlation =", ET_corr.iloc[0,1])
```

```python
# plot the data with the regression line
fig, ax = plt.subplots()
ax.scatter(eliTemp['ELI'], eliTemp['temperature'])
ax.plot(eliTemp['ELI'], m * eliTemp['ELI'] + b)
ax.set_xlabel("ELI")
ax.set_ylabel("South Central IN Min Temp Dec-Feb [F]")
plt.show()
```

The correlation between winter minimum temperature in South Central Indiana and ELI is not good, with a coefficient of -0.11. However, this is more correlation than what was expected from the analysis, so I may have done something different with the data. 

From this analysis, we can conclude that El Nino may have a small (or no) effect on these winter temperatures. The coefficient is just above what is considered negligible (<0.1).
